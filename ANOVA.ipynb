```python
import pandas as pd
import statsmodels.api as sm
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from statsmodels.formula.api import ols
```

```python
file_path = '/content/IMB737-XLS-ENG (1).xlsx'
sheet_name = 'Regression Data'
data = pd.read_excel(file_path, sheet_name=sheet_name)
```

```python
descriptive_stats = pd.DataFrame({
    'Mean': data.mean(),
    'Median': data.median(),
    'Mode': data.mode().iloc[0]
})
print("Descriptive Statistics (Mean, Median, Mode):")
print(descriptive_stats)
```

```python
IPA = ols('income_per_acre ~ loan_amount+loan_repaid+ crop_insured+ training_on_sericulture+ krishi_pond+ borewell_recharge+ rain_harvesting + own_compost_manure + own_vermi_compost + years_of_exp_in_sericulture',data=data).fit()
print("\nðŸ”¹ Regression Model Summary:")
print(IPA.summary())
```

```python
aov= sm.stats.anova_lm(IPA, typ =1)
print(aov)
```
